# Imports

```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import locale
```

```python
#view data from year 2000
df_2000 = pd.read_csv("Data/final_tmdb_data_2000.csv.gz")
print(df_2000.info())
df_2000.head()
```

```python
#view data from year 2001
df_2001 = pd.read_csv("Data/final_tmdb_data_2001.csv.gz")
print(df_2001.info())
df_2001.head()
```

# Exploratory Data Analysis (EDA)

```python
# drop first row with no data
df_2000.drop(0, inplace = True)
df_2001.drop(0, inplace = True)

# concatenate dataframes
new_df = pd.concat([df_2000, df_2001], ignore_index = True)
print(new_df.info())
new_df.head()
```

How many movies had at least some valid financial information (values > 0 for budget OR revenue)?

```python
df = new_df[(new_df['budget'] > 0) | (new_df['revenue'] > 0)]
print(f" There are {len(df)} movies with valid financial information.")
```

How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
# use value_counts to count number of movies in each certification category
categories = df.certification.value_counts()

# loop through series to print
for index, value in categories.items():
    
    # restrict to categories asked for
    if index in ['G', 'PG', 'PG-13', 'R', 'NR']:
        print(f"There are {value} movies rated {index}.")
```

What is the average revenue per certification category?

```python
# use groupby to get revenue for each certification category
revenue_per_certification = df.groupby('certification')['revenue'].mean()

# loop through series to print
for index, value in revenue_per_certification.items():
    
    # restrict to categories asked for
    if index in ['G', 'PG', 'PG-13', 'R', 'NR']:
        
        # format numbers: 3 decimal places and commas
        formatted_value = locale.format_string('%.3f', value, grouping = True)
        print(f"The average revenue of {index}-rated movies is ${formatted_value}.")
```

What is the average budget per certification category?

```python
# use groupby to get budget for each certification category
budget_per_certification = df.groupby('certification')['budget'].mean()

# loop through series to print
for index, value in budget_per_certification.items():
    
    # restrict to categories asked for
    if index in ['G', 'PG', 'PG-13', 'R']:
        
        # format numbers: 2 decimal places and commas
        formatted_value = locale.format_string('%.2f', value, grouping = True)
        print(f"The average budget of {index}-rated movies is ${formatted_value}.")
```
